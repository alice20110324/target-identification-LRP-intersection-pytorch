```python
#开始  -----方案二-----------最重要的一段代码-------------------#截止

from sklearn.datasets import make_blobs
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
#from data_process import preData,getFeaturesData,getDF,convert_to_df
import os 
import itertools
#file_dir="data/guan/"
file_dir="dataset/qiuguan/results_K_FOLD_800/gene_100/"
f_mlp_name=os.path.join(file_dir,'NFM_layer_3_encode_100_10_genes.csv')
mlp_df=pd.read_csv(f_mlp_name,sep=',')
#print(mlp_df)

mlp_df=mlp_df.iloc[:,1:]

f_nfm_3_name=os.path.join(file_dir,'NFM_layer_3_encode_1000_10_genes.csv')
nfm_3_df=pd.read_csv(f_nfm_3_name,sep=',')
nfm_3_df=nfm_3_df.iloc[:,1:]

genes_all=[]
genes_mlp=[[]]*9
genes_nfm_3=[[]]*9
genes=[[]]*9
for i in range(9):
    genes_mlp[i]=[gene for gene in mlp_df.iloc[i,:].tolist()]
    genes_nfm_3[i]=[gene for gene in nfm_3_df.iloc[i,:].tolist()]
    
    genes[i]=[gene for gene in genes_mlp[i] if gene  in genes_nfm_3[i]]
    genes_all.append(genes[i]) 
    print(len(genes[i]))
#print(genes_all)

s=0
for i in range(9):
    size=len(genes_all[i])
    s+=size
print(s)


import itertools

genes_names_for_all = list(itertools.chain(*genes_all))
genes_names_for_all.append('label')
print(len(genes_names_for_all) ) 

genes_names_for_all_df=pd.DataFrame(genes_names_for_all)
genes_names_for_all_df.to_csv('dataset/qiuguan/results_K_FOLD_800/genes_for_reconstruct/genes_100/MLP_non_encode/genes_names_for_all_df.csv')


#到此处截止

genes_list=[0.0001 for i in range(s+1)]
genes_dis_array=np.array([genes_list]*9)

print(genes_dis_array.shape)
#print(genes_dis_array)

genes_dis_array_df=pd.DataFrame(genes_dis_array)

genes_dis_array_df.columns=genes_names_for_all

#genes_dis_array_df.iloc[:,-1]=genes_dis_array_df.iloc[:,-1].astype(int)
genes_dis_array_df['label']=genes_dis_array_df['label'].astype('int')
print(genes_dis_array_df)




for i,label in enumerate(genes_dis_array_df.iloc[:,-1]):
    label=i
    genes_dis_array_df.iloc[i,-1]=label
#genes_dis_array_df.iloc[:,-1]=genes_dis_array_df.iloc[:,-1].astype(int)
print(genes_dis_array_df)

#print(genes_names_for_every_dis[1])
rows_name=genes_dis_array_df.index=['Con(0)','DN(1)','FSGS(2)','HT(3)','IgA(4)','MCD(5)','MN(6)','RPGN(7)','SLE(8)']

print(genes_dis_array_df)



for i in range(9):
    genes_dis_array_df.loc[rows_name[i],genes_all[i]]=1


print(genes_dis_array_df)

genes_dis_array_df=pd.DataFrame(genes_dis_array_df)
genes_dis_array_df.to_csv('dataset/qiuguan/results_K_FOLD_800/genes_for_reconstruct/genes_100/MLP_non_encode/genes_dis_array_df.csv')
```
