```python
import pandas as pd    
import numpy as np


#go_50='dataset/qiuguan/origin_800/LRP_aug/10/gene_analysis/F10_genes/gene_enrichment.csv'
#go_50='dataset/qiuguan/origin_800/LRP_aug/10/gene_analysis/F10_genes/BP.csv'
#go_50='dataset/qiuguan/origin_800/LRP_aug/10/gene_analysis/F10_genes/CC.csv'
go_50='dataset/qiuguan/origin_800/LRP_aug/10/gene_analysis/F10_genes/BP_MF_PATHWAY.csv'
go_50_df=pd.read_csv(go_50,sep=',')
print(go_50_df)
go_50_df.rename(columns={'Category': 'Caty'}, inplace=True)
print(go_50_df)
```

```python
genes=go_50_df.loc[:,'Genes']
caties=go_50_df.loc[:,'Caty']

print(genes)
print(caties)
```

```python
row,col=go_50_df.shape

genes_dict={}

caty_dict={}
BP_num=0
CC_num=0
MF_num=0
KEGG_num=0

gene_dict={}
for i in range(row):
    gene=genes.iloc[i]
    '''
    caty=caties.iloc[i]
    
    if caty=='BP':
        BP_num=BP_num+1
    if caty=='CC':
        CC_num=CC_num+1
    if caty=='MF':
        MF_num=MF_num+1
    if caty=='KEGG_PATHWAY':
        KEGG_num=KEGG_num+1
    '''
    
    #print(gene)
    sGene=gene.split(',')
    #print(sGene)
    
    for j in sGene:
        j=j.strip()##########一定要去掉空格
        keys=gene_dict.keys()
        keys=[m for m in keys]
        #print(keys)
        if j not in keys:
            caty_dict={}
            gene_list=[]
            caty_total=0
        
            BP_num=0
            CC_num=0
            MF_num=0
            KEGG_num=0
            caty_dict['BP']=BP_num
            caty_dict['CC']=CC_num
            caty_dict['MF']=MF_num
            caty_dict['KEGG']=KEGG_num
            caty_dict['total']=caty_total

            '''  
            if caty=='BP':
                caty_dict[caty]=BP_num
            if caty=='CC':
                caty_dict[caty]=CC_num
            if caty=='MF':
                caty_dict[caty]=MF_num
            if caty=='KEGG_PATHWAY':
                caty_dict[caty]=KEGG_num
        
            '''
        
        
            for k in range(row):
                g=genes.iloc[k]
                if g.find(j)>=0:
                    #print('gene:',j)
                    #print("g:",g)
                
                    caty=caties.iloc[k]
                    #print('caty:',caty)
                    if caty=='BP':
                        caty_dict[caty]=BP_num+1
                        caty_total=caty_total+1
                        #caty_dict['total']=caty_total
                    if caty=='CC':
                        caty_dict[caty]=CC_num+1
                        caty_total=caty_total+1
                        #caty_dict['total']=caty_total
                    if caty=='MF':
                        caty_dict[caty]=MF_num+1
                        caty_total=caty_total+1
                        #caty_dict['total']=caty_total
                    if caty=='KEGG_PATHWAY':
                        caty_dict[caty]=KEGG_num+1
                        caty_total=caty_total+1
                        #caty_dict['total']=caty_total
                    caty_dict['total']=caty_total
                
            gene_dict[j]=caty_dict
                
print(gene_dict)                
```
