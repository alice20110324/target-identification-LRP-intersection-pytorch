```python
#文章知识点与官方知识档案匹配，可进一步学习相关知识
#————————————————
#版权声明：本文为CSDN博主「李划水员」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
#原文链接：https://blog.csdn.net/Daisy4/article/details/124291254
import numpy as np
import pandas as pd
import csv
txt_file='dataset/qiuguan/orign/xiaoqiu.txt'
out = open('dataset/qiuguan/orign/T/qiu.csv','w',newline='')        #要转成的.csv文件，先创建一个LF1big.csv文件
csv_writer = csv.writer(out,dialect='excel')  
 
f = open(txt_file,"r")
for line in f.readlines():
    line=line.replace(',','\t')  #将每行的逗号替换成空格
    list = line.split()          #将字符串转为列表，从而可以按单元格写入csv
    csv_writer.writerow(list)
```

```python
df=pd.read_csv('dataset/qiuguan/orign/T/qiu.csv',sep=',',header=None)
#print(df)
df=df.T
#df=df.iloc[1:,1:]
df.to_csv('dataset/qiuguan/orign/T/qiu.csv')
#print(df)
```
