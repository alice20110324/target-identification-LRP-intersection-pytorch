```python
#合并各种疾病文件
import pandas as pd 
import numpy as np
#programm test
import os
file_dir="dataset/qiuguan/results_K_FOLD_800/final_genes/"
all_file_list=os.listdir(file_dir)
#print(all_file_list)
n=0
genes_names_for_every_dis=[]
genes_names_for_all=[]
for single_file in all_file_list:
    #print(single_file)
    df1=pd.read_csv(file_dir+single_file,sep=',')
    df1_1=df1.iloc[:,1:]
    rows,cols=df1_1.shape
    df1_1=df1_1.values.reshape(rows).tolist()
    
    #printnt()
    #print(df1.iloc[,:])
    #print(df1_1)
    #prnt()
    n=n+1
    genes_names_for_every_dis.append(df1_1)
    genes_names_for_all.append(df1_1)

#拼接list中的array，去掉arry的[]，全部变成list    
import itertools

genes_names_for_all = list(itertools.chain(*genes_names_for_all))
print(len(genes_names_for_all) ) 


genes_names_for_all = list(set(genes_names_for_all))
genes_names_for_all.append('label')
print(len(genes_names_for_all))



genes_names_for_all_df=pd.DataFrame(genes_names_for_all)
genes_names_for_all_df.to_csv('dataset/qiuguan/results_K_FOLD_800/genes_for_reconstruct/genes_names_for_all_df.csv')
cols_num=len(genes_names_for_all)

#构建疾病基因矩阵，行为疾病，列为基因

genes_list=[0.0001 for i in range(cols_num)]
genes_dis_array=np.array([genes_list]*9)

print(genes_dis_array.shape)
#print(genes_dis_array)

genes_dis_array_df=pd.DataFrame(genes_dis_array)

genes_dis_array_df.columns=genes_names_for_all

genes_dis_array_df.iloc[:,-1].astype(int)
#print(genes_dis_array_df)

for i,label in enumerate(genes_dis_array_df.iloc[:,-1]):
    label=i
    genes_dis_array_df.iloc[i,-1]=label
genes_dis_array_df.iloc[:,-1]=genes_dis_array_df.iloc[:,-1].astype(int)
#print(genes_dis_array_df)

#print(genes_names_for_every_dis[1])
rows_name=genes_dis_array_df.index=['Con(0)','DN(1)','FSGS(2)','HT(3)','IgA(4)','MCD(5)','MN(6)','RPGN(7)','SLE(8)']

print(genes_dis_array_df)



for i in range(9):
    genes_dis_array_df.loc[rows_name[i],genes_names_for_every_dis[i]]=1


print(genes_dis_array_df)

genes_dis_array_df=pd.DataFrame(genes_dis_array_df)
genes_dis_array_df.to_csv('dataset/qiuguan/results_K_FOLD_800/genes_for_reconstruct/genes_dis_array_df.csv')
```
